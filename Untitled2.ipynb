```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *
```

```python
pagespeed = np.random.normal(3.0, 1.0, 1000)
purchaseamount = 100-(pagespeed + np.random.normal(0.0, 0.1, 1000))*3
scatter(pagespeed, purchaseamount)
```

```python
from scipy import stats 
slope, intercept, r_value, p_value, std_error = stats.linregress(pagespeed, purchaseamount)
```

```python
print(slope)
print(intercept)
print(r_value)
print(p_value)
print(std_error)
```
